#**Qus 10 Please create a new column named total_bedroom_size. If the size is 10 or less, it should be quoted as small. If the size is 11 or more but less than 1000, it should be medium, otherwise, it should be considered large**

```python
def total_bedroom_size(column): #creating a function name bedrooms_size having variable inside coloum
  if column <=10: # appling a logic   If the size is 10 or less 
    return "small" # it return the small
  elif column > 10 or column <1000: #appling a logic If the size is 11 or more but less than 1000
    return "medium" #it show return medium
  else :  
    return "large" #it show large
```

```python
new_df["total_bedroom_size"] = new_df["total_bedrooms"].apply(total_bedroom_size) #this code used great a new coloum having name 
# (total_bedroom_size) and store the value also. and in population colume the self define function (bedroom_size apply) 
```
